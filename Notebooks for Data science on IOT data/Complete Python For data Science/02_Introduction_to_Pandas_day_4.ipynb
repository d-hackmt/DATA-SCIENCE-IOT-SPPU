What is Data?

Data refers to facts or information used for analysis or decision-making. It can take many forms, such as numbers, text, images, or even sounds.
Example: A table of student names, their ages, and grades is data.

1D Data (One-Dimensional):
Represents a single sequence of data, like a list or series.
Example: A list of student names or a single column of text.

2D Data (Two-Dimensional):
Organized in rows and columns, like tables or spreadsheets.
Example: A DataFrame, an Excel sheet, or a collection of grayscale (B&W) images.

3D Data (Three-Dimensional):
Data with layers, often representing frames or sequences.
Example: A video where each frame is a 2D image stacked over time.

4D Data (Four-Dimensional):
Adds another dimension, such as sound or metadata, to a 3D structure.
Example: A video with audio or temporal changes in medical imaging (e.g., MRI scans over time).

ND Data (N-Dimensional):
Extends beyond 4D and is used in specialized fields.
Example: Data from scientific simulations or 3D medical scans with time and additional variables.

Why Use a Series?

1. Simplifies data manipulation and analysis for 1D data.

### DataFrames In Pandas

What is a Pandas DataFrame?

1. A DataFrame is a two-dimensional, size-mutable, and heterogeneous data structure with labeled axes (rows and columns).
2. It is one of the most widely used tools in Pandas for data manipulation and analysis.

Think of it as a spreadsheet or an SQL table, but more powerful and flexible.
A DataFrame allows you to organize data into rows and columns, perform operations efficiently, and visualize tabular data.


Key Features of a DataFrame
* Heterogeneous Data: Can store different types of data (integers, floats,
strings, etc.) in one table.
* Labeled Axes: Rows and columns have labels that make accessing and manipulating data intuitive.
* Flexible Input: DataFrames can be created from dictionaries, lists, arrays, Series, or even files like CSV and Excel.

Creating DataFrames

1. Creating a DataFrame from a Dictionary of Lists
This is one of the most common ways to create a DataFrame.

````python
import pandas as pd

# Example: Employee details
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Salary': [50000, 60000, 70000]
}
df = pd.DataFrame(data)
print("DataFrame from Dictionary of Lists:\n", df)
````

### Indexing in Pandas


image.png

What is Indexing?

* Indexing in Pandas is a way to select and access data in a DataFrame or Series using labels (row/column names) or positions (row/column numbers). It is one of the most fundamental operations for data manipulation and analysis.


Why is Indexing Important?

* It allows you to efficiently locate, filter, and manipulate specific parts of the data.
Helps in slicing, filtering, and performing operations on subsets of data.


----------------
Indexing Methods

1. Label-Based Indexing (loc)


The loc method selects data based on row and column labels. It is intuitive and works with labeled axes.


Syntax
```
df.loc[row_labels, column_labels]

```



````python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Salary': [50000, 60000, 70000, 80000]
}
df = pd.DataFrame(data, index=['a', 'b', 'c', 'd'])

# Access a single row by label
print(df.loc['a'])

print("                ")
print("----------------")

# Access multiple rows
print(df.loc[['a', 'c']])

print("                ")
print("----------------")

# Access a specific column
print(df.loc[:, 'Age'])

print("                ")
print("----------------")

# Access specific rows and columns
print(df.loc[['a', 'd'], ['Name', 'Salary']])
````

 -

2. Position-Based Indexing (iloc)

* The iloc method selects data based on integer positions.
* It is useful when labels are not known or when working with large datasets.


Syntax
```
df.iloc[row_positions, column_positions]

```



````python
# Access the first row
print(df.iloc[0])

print("                ")
print("----------------")

# Access the first two rows
print(df.iloc[0:2])

print("                ")
print("----------------")

# Access a specific cell (first row, second column)
print(df.iloc[0, 1])


print("                ")
print("----------------")

# Access all rows and the first two columns
print(df.iloc[:, 0:2])

````

### Sorting and Ranking Data: Basics

Sorting and ranking are essential operations in data analysis, enabling structured and meaningful insights from large datasets.

They help organize data systematically and highlight the most important or relevant information for decision-making.





What is Sorting?

Sorting is the process of arranging data in a particular order. It can be based on the values in one or more columns and can be ascending or descending. Sorting helps to:

Identify trends (e.g., highest or lowest sales).
Compare values across categories.
Prepare data for further analysis.

Types of Sorting in Pandas:

Sort by Values: Sort rows or columns based on their values.

Sort by Index: Sort the data based on the index.

What is Ranking?

Ranking assigns a rank to each value in a column relative to other values. It's particularly useful when you want to determine relative positions (e.g., top 3 performers).


It handles ties by assigning the same rank to duplicate values and skipping ranks accordingly.
Ranks can be assigned in ascending or descending order.


Examples and Applications

--------

* Example 1: Sorting by Single Column


Application: Sorting employees by age to analyze workforce demographics.



````python
import pandas as pd

# Data
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40],
        'Salary': [50000, 60000, 70000, 80000]}
df = pd.DataFrame(data)

# Sort by 'Age'
sorted_age = df.sort_values(by='Age')
print("Sorted by Age:\n", sorted_age)
````

* Example 2: Sorting by Multiple Columns

Application: Finding the highest-earning employees grouped by age categories.

````python
# Sort by 'Salary' (descending), then by 'Age' (ascending)
sorted_salary_age = df.sort_values(by=['Salary', 'Age'], ascending=[False, True])
print("Sorted by Salary (desc) and Age (asc):\n", sorted_salary_age)
````

Example 3: Sorting by Index

Application: Reorganizing data for better readability and alignment with other datasets.

````python
# Sort by index
sorted_index = df.sort_index()
print("Sorted by Index:\n", sorted_index)
````

Example 4: Ranking Based on Column Values

Application: Ranking players by scores or ranking products by popularity.

````python
# Add a 'Salary Rank' column
df['Salary_Rank'] = df['Salary'].rank(ascending=False)
print("Ranked by Salary:\n", df)
````

Example 5: Using Sorting and Ranking Together

Application: Prioritizing employee appraisals based on age and salary.|

````python
# Sorting by Age and then ranking by Salary
sorted_df = df.sort_values(by='Age')
sorted_df['Salary_Rank'] = sorted_df['Salary'].rank(ascending=False)
print("Sorted by Age and Ranked by Salary:\n", sorted_df)
````

### Handling Missing Data: `fillna`,` dropna`, `isna`

In real-world datasets, missing data is common. It occurs due to errors in data collection, storage, or transmission. Handling missing data is crucial because:

1. Missing values can lead to errors in analysis.

2. They can mislead results if not treated properly.

There are three main ways to handle missing data:

* Remove missing data: Use dropna().

* Fill missing data: Use fillna() with appropriate values (mean, median, mode, or custom values).

* Identify missing data: Use isna() or notna() to check for missing values.

Handling missing data is one of the most crucial tasks in data preprocessing.

 Missing values are common in real-world datasets, and improper handling can significantly affect the results of data analysis or machine learning models.


Pandas provides multiple ways to deal with missing data, such as fillna(), dropna(), and isna().

Common Causes of Missing Data

* Data Collection Issues: Some data might not be collected due to errors or lack of access.

* Data Corruption: Incomplete data entry or transmission errors can introduce null values.

* Non-applicable Data: Sometimes, a value might not apply to a specific record (e.g., a missing age for an organization).

Types of Missing Data

1. NaN (Not a Number): Used for numerical columns where data is missing.

2. None: Used for object (string) columns.

Step 1: Creating a DataFrame with Missing Values

Lets start with a sample DataFrame containing some missing values.

````python
import pandas as pd
import numpy as np

# Creating a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Helen'],
    'Age': [25, np.nan, 30, 35, 40, np.nan, 28, 33],
    'Salary': [50000, 60000, 55000, np.nan, 45000, 48000, np.nan, 52000],
    'Department': ['HR', 'Finance', np.nan, 'IT', 'HR', 'Finance', 'IT', 'HR']
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
````

Step 2: Handling Missing Data

Option 1: Dropping Missing Data (dropna())

Removes rows or columns with missing values.

Use when missing values are few, and dropping them doesnt impact analysis.

````python
# Dropping rows with missing values
df_dropped_rows = df.dropna()
print("\nDataFrame after dropping rows with missing values:")
print(df_dropped_rows)

# Dropping columns with missing values
df_dropped_cols = df.dropna(axis=1)
print("\nDataFrame after dropping columns with missing values:")
print(df_dropped_cols)
````

Option 2: Filling Missing Data (fillna())
Fill missing values with specific values like mean, median, or mode.

a) Fill with Mean
Use when data is numerical and has a normal distribution.

````python
# Filling missing values in 'Age' column with mean
mean_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean_age)
print("\nDataFrame after filling 'Age' with mean value:")
print(df)
````

b) Fill with Median
Use when data is numerical and has outliers.

````python
# Filling missing values in 'Salary' column with median
median_salary = df['Salary'].median()
df['Salary'] = df['Salary'].fillna(median_salary)
print("\nDataFrame after filling 'Salary' with median value:")
print(df)
````

c) Fill with Mode
Use when data is categorical.



````python
# Filling missing values in 'Department' column with mode
mode_department = df['Department'].mode()[0]
df['Department'] = df['Department'].fillna(mode_department)
print("\nDataFrame after filling 'Department' with mode value:")
print(df)
````

---------

Option 3: Identifying Missing Data (isna(), notna())
Check for missing values and count them.

````python
# Identifying missing values
print("\nChecking for missing values:")
print(df.isna())

# Counting missing values per column
missing_values_count = df.isna().sum()
print("\nCount of missing values per column:")
print(missing_values_count)
````

--


Applications of Handling Missing Data in Real Life

1. Finance:

Incomplete financial records can be imputed with mean or median to ensure accurate financial reporting and predictions.

2. Healthcare:

Missing patient data (e.g., age, weight, or blood pressure) can be filled with median values to preserve the integrity of datasets.

3. E-Commerce:

Product ratings might have missing values. Imputing with mode (most frequent rating) can
maintain consistent reviews.

4. Weather Data:

Missing temperature or precipitation data can be filled using the median or interpolated values to maintain data continuity.

### Writing to Files (CSV, Excel, JSON)


In data analysis, the ability to save and export data in various formats is crucial for sharing, reporting, and storing processed datasets. Pandas provides seamless functionality to write DataFrames to file formats such as CSV, Excel, and JSON, making it a powerful tool for handling data pipelines.



Why Writing Files is Important?

Data Sharing: Allows collaboration by exporting data for team members or stakeholders.

Persistence: Saves the processed data for future use or archival purposes.

Interoperability: Exports data into formats compatible with other tools and systems.

1. Writing CSV Files

CSV is one of the most commonly used formats for saving tabular data due to its simplicity and wide compatibility.

Syntax to Write CSV Files



```
df.to_csv('file_name.csv', index=False)

```



Example: Writing a CSV File

First, create a DataFrame:

````python
import pandas as pd

# Creating a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Salary': [50000, 60000, 70000]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Writing to a CSV file
df.to_csv('employees.csv', index=False)
print("\nCSV file 'employees.csv' has been created!")
````

Customizing CSV Writing

Include Index: Save with row indices.



````python
df.to_csv('employees_with_index.csv', index=True)
````

Delimiter: Use a custom delimiter (e.g., tab-separated)

````python
df.to_csv('employees. tsv', sep='\t', index=False)
````

2. Writing Excel Files

Excel files are versatile for business reporting and data sharing.

Syntax to Write Excel Files



```
df.to_excel('file_name.xlsx', sheet_name='Sheet1', index=False)

```



Example: Writing an Excel File

Create and save a DataFrame:

````python
# Creating a DataFrame
data = {
    'Product': ['Laptop', 'Mouse', 'Keyboard'],
    'Price': [1000, 50, 80],
    'Stock': [50, 200, 150]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Writing to an Excel file
df.to_excel('inventory.xlsx', sheet_name='Inventory', index=False)
print("\nExcel file 'inventory.xlsx' has been created!")
````

Customizing Excel Writing
Multiple Sheets: Save different DataFrames to multiple sheets.

````python
with pd.ExcelWriter('multi_sheet.xlsx') as writer:
    df.to_excel(writer, sheet_name='Sheet1', index=False)
    df.to_excel(writer, sheet_name='Sheet2', index=False)
````

3. Writing JSON Files

JSON is widely used for structured data storage and web communication.

Syntax to Write JSON Files



```
df.to_json('file_name.json', orient='records', indent=4)
```




Example: Writing a JSON File

Create and save a DataFrame:

````python
# Creating a DataFrame
data = {
    'Country': ['USA', 'India', 'Germany'],
    'Capital': ['Washington D.C.', 'New Delhi', 'Berlin'],
    'Population': [331, 1380, 83]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Writing to a JSON file
df.to_json('countries.json', orient='records', indent=4)
print("\nJSON file 'countries.json' has been created!")
````

Customizing JSON Writing

Different Orientations:

````python
df.to_json('countries_split.json', orient='split', indent=4)
````

Practical Demonstration: Combining All Formats

Step 1: Creating a Large Dataset




````python
import numpy as np

# Creating a large dataset
data = {
    'ID': range(1, 21),
    'Name': [f'User{i}' for i in range(1, 21)],
    'Age': np.random.randint(20, 50, 20),
    'Score': np.random.uniform(50, 100, 20)
}
df_large = pd.DataFrame(data)
print("Large Dataset:")
print(df_large)
````

Step 2: Saving to All Formats

````python
# Writing to CSV
df_large.to_csv('large_dataset.csv', index=False)

# Writing to Excel
df_large.to_excel('large_dataset.xlsx', sheet_name='Data', index=False)

# Writing to JSON
df_large.to_json('large_dataset.json', orient='records', indent=4)
print("\nFiles 'large_dataset.csv', 'large_dataset.xlsx', and 'large_dataset.json' have been created!")
````
